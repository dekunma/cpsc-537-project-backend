```python
import pandas as pd
```

```python

# read csv
mbti_movie = pd.read_csv('./mbti_movie.csv')
mbti_100 = pd.read_csv('./mbti_100.csv')
```

```python
mbti_100['name'] = mbti_100['name'].str.replace("'", "''")
mbti_100['name'] = mbti_100['name'].str.replace("’", "''")
mbti_100['name'] = mbti_100['name'].str.replace("“", '"')
mbti_100['name'] = mbti_100['name'].str.replace("”", '"')
mbti_100['personality'] = mbti_100['personality'].str.upper()
mbti_100['description'] = mbti_100['name'].apply(lambda x: x[x.find("(")+1:x.find(")")].strip() if "(" in x else "")
mbti_100['name'] = mbti_100['name'].apply(lambda x: x[:x.find("(")].strip() if "(" in x else x.strip())
mbti_100
```

```python
mbti_movie['role'] = mbti_movie['role'].str.replace("'", "''")
mbti_movie['movie'] = mbti_movie['movie'].str.replace("'", "''")
mbti_movie['movie'] = mbti_movie['movie'].str.strip()
mbti_movie
```

```python
mbti_100_people = [(row['name'], row['personality']) for _, row in mbti_100[['name', 'personality']].iterrows()]
mbti_100_people[:10]
```

```python
mbti_movie_people = [(row['role'], row['mbti']) for _, row in mbti_movie[['role', 'mbti']].iterrows()]
mbti_movie_people[:10]
```

```python
# join 2 tuples
# in case of duplicates, use mbti_movie_people
name2mbti = dict(mbti_movie_people)
for name, mbti in mbti_100_people:
    if name not in name2mbti:
        name2mbti[name] = mbti
    
for i, (name, mbti) in enumerate(name2mbti.items()):
    if i == 10:
        break
    print([name, mbti])
```

```python
# person table
output_file = open('./person_table.txt', 'w')
output_file.write('DROP TABLE IF EXISTS person;')
output_file.write('CREATE TABLE person(name VARCHAR(512) PRIMARY KEY,code VARCHAR(4));')

for name, mbti in name2mbti.items():
    print(f'INSERT IGNORE INTO person (name, code) VALUES (\'{name}\', \'{mbti}\');', file=output_file)
```

```python
# mbti_movie table
output_file = open('./mbti_movie_table.txt', 'w')
output_file.write('DROP TABLE IF EXISTS mbti_movie;')
output_file.write('CREATE TABLE mbti_movie(name	VARCHAR(512) PRIMARY KEY,movie VARCHAR(512));')

for name, movie in mbti_movie[['role', 'movie']].values:
    print(f'INSERT IGNORE INTO mbti_movie (name, movie) VALUES (\'{name}\', \'{movie}\');', file=output_file)
```

```python
# mbti_100 table
output_file = open('./mbti_100_table.txt', 'w')
output_file.write('DROP TABLE IF EXISTS mbti_100;')
output_file.write('CREATE TABLE mbti_100 (name	VARCHAR(512) PRIMARY KEY, description VARCHAR(512), total_votes INT,first_letter_votes_percentage	VARCHAR(512),second_letter_votes_percentage	VARCHAR(512),third_letter_votes_percentage	VARCHAR(512),fourth_letter_votes_percentage	VARCHAR(512));')

for name, description, total_votes, first_letter_votes_percentage, second_letter_votes_percentage, third_letter_votes_percentage, fourth_letter_votes_percentage in mbti_100[['name', 'description', 'total votes', 'first letter votes percentage', 'second letter votes percentage', 'third letter votes percentage', 'fourth letter votes percentage']].values:
    print(f'INSERT IGNORE INTO mbti_100 (name, description, total_votes, first_letter_votes_percentage, second_letter_votes_percentage, third_letter_votes_percentage, fourth_letter_votes_percentage) VALUES (\'{name}\', \'{description}\', {total_votes}, \'{first_letter_votes_percentage}\', \'{second_letter_votes_percentage}\', \'{third_letter_votes_percentage}\', \'{fourth_letter_votes_percentage}\');', file=output_file)
```
